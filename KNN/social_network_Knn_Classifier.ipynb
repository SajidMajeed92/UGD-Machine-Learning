## Social Network Advertisment
Dataset on Social media ads describing users, whether users have purchased a product by clicking on the advertisements shown to them.

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/Social_Network_Ads.csv')
df.head()
```

```python
X=df.iloc[:,2:4]
```

```python
X
```

```python
X.shape
```

```python
y = df.Purchased.values
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
X_test
```

```python
# 1st method of finding k
np.sqrt(X_train.shape[0])
```

```python
k = 17
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# Effect of K on Training Error rate and Validation Error rate

error_train=[]
error_test=[]

for i in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    x=confusion_matrix(y_train,knn.predict(X_train))
    y=confusion_matrix(y_test,knn.predict(X_test))
    error_train.append((x[0][1]+x[1][0])/x.sum())
    error_test.append((y[0][1]+y[1][0])/y.sum())
```

```python
plt.plot(range(1,26),error_train,label='training error rate')
plt.plot(range(1,26),error_test,label='test/validation error rate')
plt.xlabel('K Value')
plt.ylabel('Error')
plt.legend()
```

```python
# 2nd method to find the value of K
accuracy = []

for i in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    accuracy.append(accuracy_score(y_test,knn.predict(X_test)))
```

```python
accuracy
```

```python
np.argmax(accuracy)
```

```python
plt.plot(range(1,26),accuracy)
```
